```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np

import ast
```

```python
movies = pd.read_csv('/content/drive/MyDrive/Movie Dataset/tmdb_5000_movies.csv')
credits = pd.read_csv('/content/drive/MyDrive/Movie Dataset/tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies.info()
```

```python
#taking only gener id keywords title overview cast crew
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.info()
```

```python
#converting dictionary to list
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
#taking only first three actors
movies['cast'] = movies['cast'].apply(lambda x: x[:3])
movies.head()
```

```python
# Function to fetch director names
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
#taking only directors as a crew
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head()
```

```python
#removing space
def collapse(L):
    L1 = [i.replace(" ", "") for i in L]
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
# Combined the tags into a single string
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))
movies.head()
```

```python
movies.head()
```

```python
# Convert text data to a numerical format using TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(movies['tags'])
```

```python
# Calculate similarity matrix using cosine similarity
similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)
np.fill_diagonal(similarity_matrix, 0)
```

```python
# Use DBSCAN for clustering
dbscan = DBSCAN(eps=0.001, min_samples=500, metric='precomputed')
dbscan.fit(similarity_matrix)
```

```python
# Add cluster labels to the movies DataFrame
movies['cluster'] = dbscan.labels_
```

```python
# Function to recommend movies using DBSCAN
def recommend_dbscan(movie):
    movie_index = movies[movies['title'] == movie].index[0]
    cluster_label = movies.iloc[movie_index]['cluster']

    # Extract movies from the same cluster
    cluster_movies = movies[movies['cluster'] == cluster_label]

    # Exclude the input movie from recommendations
    cluster_movies = cluster_movies[cluster_movies['title'] != movie]

    # Randomly shuffle the cluster movies for variety
    cluster_movies = cluster_movies.sample(frac=1)

    # Display recommended movies from the same cluster
    recommended_movies = cluster_movies['title'][:5].tolist()

    return recommended_movies
```

```python
def recommend_movies(movie):
    recommend_movies = recommend_dbscan(movie)
    print("Recommended Movies using DBSCAN:")
    for recommended_movie in recommend_movies:
        print(recommended_movie)
```
